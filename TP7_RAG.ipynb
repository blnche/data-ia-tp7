# **TP7: Retrieval-Augmented Generation (RAG)** <br> A Case Study with Sanofi's Annual Report 2022
---

Le rapport annuel de Sanofi est un document détaillé qui présente les performances financières, les réalisations et les objectifs de l'entreprise pour l'année écoulée. Il est utilisé par les investisseurs, les analystes financiers et les parties prenantes pour évaluer la santé financière de l'entreprise et sa stratégie future.

Dans le cadre de ce TP, nous allons utiliser le modèle RAG (Retrieval-Augmented Generation) pour répondre à des questions sur le [rapport annuel de Sanofi de 2022](https://www.sanofi.com/assets/dotcom/content-app/publications/integrated-report/SANOFI-Integrated-Annual-Report-2022-EN.pdf) qui fait 43 pages. Le modèle RAG combine la recherche d'informations et la génération de texte pour fournir des réponses précises et contextuelles aux questions posées.

## **1. Introduction**

- Dans ce TP, nous allons voir comment utiliser le modèle RAG (Retrieval-Augmented Generation) pour répondre à des questions sur un document PDF.

<img src="./images/tp7_pdf_extractors.png">

[What’s the Best PDF Extractor for RAG? I Tried LlamaParse, Unstructured and Vectorize](https://medium.com/gitconnected/whats-the-best-pdf-extractor-for-rag-i-tried-llamaparse-unstructured-and-vectorize-4abbd57b06e0)

## **2. Contexte**
- Votre client Sanofi souhaite mettre en place un chatbot pour répondre aux questions des utilisateurs sur le rendement annuel de l'entreprise.
- Vous avez à votre disposition le rapport annuel de l'entreprise au format PDF.
- Vous avez choisi d'utiliser le modèle RAG pour répondre à ces questions.

## **3. Objectifs**
- Développer une architecture RAG avec des prompts efficaces pour répondre à ces questions :
    - Quels sont les objectifs environnementaux de Sanofi en matière de neutralité carbone et quelles actions concrètes ont été entreprises en 2022 pour y parvenir ?
    - Quelles avancées majeures ont été réalisées dans le développement du médicament Dupixent® en 2022 et pour quelles nouvelles indications a-t-il été approuvé ?
    - Quelle est la mission de la fondation "Foundation S – The Sanofi Collective", et quels résultats concrets a-t-elle obtenus en 2022 ?
    - Comment Sanofi utilise-t-elle l’intelligence artificielle pour accélérer la recherche et le développement de nouveaux médicaments ? Donne des exemples de partenariats ou de projets spécifiques.
    - Quelles mesures Sanofi a-t-elle mises en place pour promouvoir la diversité, l'équité et l'inclusion (DE&I) dans ses effectifs et au sein de sa gouvernance en 2022 ?
    - 🚨 Quelle est la répartition des ventes de Sanofi en 2022 par zone géographique et par unité commerciale ?
